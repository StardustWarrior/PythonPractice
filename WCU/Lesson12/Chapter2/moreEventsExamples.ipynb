```python
#Updating Text
from tkinter import *

class MyFrame(Frame):
    def __init__(self):
        Frame.__init__(self)
        #Frame size
        self.master.geometry("250x200")
        #Frame title
        self.master.title("My GUI")
        self.grid()

        #Create and add the button to the frame
        self.button_click_here = Button( self, text = "Click Here", command = self.click_here_click )
        self.button_click_here.grid()

        #Control variable to update the text of the label
        self.my_text = StringVar()

        #Textvariable is a property used to update labels during runtime
        self.message = Label(self, textvariable = self.my_text)
        self.message.grid()

    def click_here_click(self):
        self.my_text.set("You did it!")

frame03 = MyFrame()
frame03.mainloop()
```

```python
#The Entry Component
from tkinter import *

class MyFrame(Frame):
    def __init__(self):
        Frame.__init__(self)
        self.master.geometry("250x200")
        self.master.title("My GUI")
        self.grid()

        self.prompt = Label(self, text = "What's your name?")
        self.prompt.grid()

        self.input = Entry(self)
        self.input.grid()

        self.button_submit = Button( self, text = "Submit", command = self.submit_click )
        self.button_submit.grid()

        self.my_text = StringVar()
        self.message = Label(self, textvariable = self.my_text)
        self.message.grid()

    def submit_click(self):
        output_message = "Hi " + self.input.get()
        self.my_text.set(output_message)

frame04 = MyFrame()
frame04.mainloop()
```

```python
#The Grid Layout Manager
#Adding Spacing Using the Grid Layout Manager
from tkinter import *

class MyFrame(Frame):
    def __init__(self):
        Frame.__init__(self)
        self.master.geometry("250x200")
        self.master.title("My GUI")
        self.grid()

        self.prompt = Label(self, text = "What's your name?")
        #Give the label where it should position, all frames have a grid layout
        self.prompt.grid(row = 0, column = 0)

        self.input = Entry(self)
        self.input.grid(row = 0, column = 1)

        self.button_submit = Button( self, text = "Submit", command = self.submit_click )
        #Add padding on axis Y to the button, 10 pixels of padding here
        self.button_submit.grid(columnspan = 2, pady = 10)

        self.my_text = StringVar()
        self.message = Label(self, textvariable = self.my_text)
        self.message.grid(columnspan = 2)

    def submit_click(self):
        output_message = "Hi " + self.input.get()
        self.my_text.set(output_message)

frame04 = MyFrame()
frame04.mainloop()
```
